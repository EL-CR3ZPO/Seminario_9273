Como dijimos antes intuitivamente pensamos que la base1 y base2 estan relacionadas y mas que eso pensamos que tal vez antes eran una sola base. Entonces para comprobar esto intentamos hacer un merge entre las bases mediante las columnas "mes_informacion" y "num_caso" y en caso de que la relación sea única el merge va a funcionar.

```python
base_comportamiento_3 = pd.merge(base1, base2, how = "right", on =['num_caso','mes_informacion'])
```

Como vimos el merge funcionó bien, ahora como no pensamos usar la columna "mes_informacion", pues aunque aporta información importante acerca de la edad del crédito, esta información la podemos obtener de la columna "Tiempo_aparicion", por lo que solo guardamos las columnas que vayamos a utilizar para que sea mas sencillo trabajar con ellas.

```python
base_apar = base_comportamiento_3[["num_caso","Tiempo_aparicion"]]
```

```python
base_apar.head()
```

```python
base_apar.info()
```

```python
base_comp = base_comportamiento_3[["num_caso","comportamiento_j"]]
```

```python
base_comp.head()
```

Ahora para resumir la información y que nos aparezcan los datos de cada cliente primero sacamos la edad del crédito para cada cliente y para ello usamos la función groupby y count para que cada que se repita el "num_caso" en la columna lo cuente y gurde en la columna "Tiempo_aparicion" que ahora se llamara "edad_credit"

```python
base_edad = base_apar.groupby(['num_caso']).count().reset_index()
renombrar={"Tiempo_aparicion":"edad_credit"}
base_edad=base_edad.rename(columns=renombrar)
```

```python
base_edad.head()
```

```python
base_edad.info()
```

En un inicio queríamos obtener el total de retraso de pagos que tuvo cada cliente, y para ello había separado la columna "comportamiento_j" para que a cada renglon a partir de la segunda posición se restara el anterior renglón para asi obtener el adeudo en caso de que lo haya en ese mes y despues guradarlo y unirlo a la base para después usar el groupby y el count y asi obtener el total de retrasos que tuvo el cliente durante el período de tiempo que tardo el liquidar el crédito, pero el proceso por lo menos para mi computadora era dificil y aunque lo estaba haciendo tardaba demasiado en terminarlo por loq eu decidí tomar el máximo de retardos consecutivos que tuvo el cliente y asi crear un citerio para otorgar otro crédito en cso de que los clientes lo requieran.

```python
base_retrasos = base_comp.groupby(['num_caso']).max().reset_index()
renombrar={"comportamiento_j":"retrasos_max"}
base_retrasos=base_retrasos.rename(columns=renombrar)
```

```python
base_retrasos.head()
```

Ya que obtuvimos los datos necesarios para nuestro primer criterio, juntamos todos los datos para poder manejarlos con mas facilidad.

```python
base_comportamiento_ = pd.merge(base_edad, base_retrasos)
```

```python
base_comportamiento_.head()
```

```python
base_comportamiento_.tail()
```

```python
base_comportamiento_.info()
```

```python
base_preliminar = base_comportamiento_[(1/3)*base_comportamiento_["edad_credit"]>base_comportamiento_["retrasos_max"]][["num_caso","edad_credit","retrasos_max"]].reset_index()
```
